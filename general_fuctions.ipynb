```python
import seaborn as sns
import pandas as pd
```

```python
# load data
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# create a data frame
titanic_df = pd.DataFrame(data=titanic)   
titanic_df.head()
```

```python
# number of columns 

titanic_df.describe()
```

```python
# select the 1st row

titanic_df.iloc[0]
```

```python
# display 4 column
titanic_df.iloc[:4]
```

```python
# extract data with Cherboug as the embark town
titanic_df = titanic_df.set_index(titanic_df['embark_town'])
titanic_df_cherboug = titanic_df.loc['Cherbourg']
titanic_df_cherboug.head()
```

```python
titanic_df_cherboug.describe()
```

```python
## Display all female with above 50 years of age 

titanic_df[(titanic_df['sex'] == 'female') & (titanic_df['age'] >= 50)]
```

```python
#  Rename Values

titanic_df_w_cls= titanic_df.rename(columns={'pclass': 'Passager Class'})
titanic_df_w_cls.head()
```

```python
titanic_df_woman= titanic_df['sex'].replace("female", "woman")
titanic_df_woman.head()
```
